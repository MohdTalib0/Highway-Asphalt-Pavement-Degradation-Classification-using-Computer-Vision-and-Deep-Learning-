```python
#code to create the yaml file for Yolov8
import yaml
data = {'train' : '/content/drive/MyDrive/omdena/Yolo/train/images', # Replace with the actual file path
'val' : '/content/drive/MyDrive/omdena/Yolo/valid/images',           # Replace with the actual file path
        'test' : '/content/drive/MyDrive/omdena/Yolo/test/images',  # Replace with the actual file path
        'names' : ['Pothole','crack','Rut','Repair']}


with open('/content/drive/MyDrive/omdena/Yolo/datatest.yaml', 'w') as outfile:
    yaml.dump(data, outfile, default_flow_style=False)
```

```python
# Train the model small yolov8s
from ultralytics import YOLO
model = YOLO('yolov8s.pt')
model.train(
   data='/content/drive/MyDrive/omdena/Yolo/datatest.yaml', # Replace with the actual file path
   epochs=25,batch=32,name='/content/drive/MyDrive/omdena/runs/detect/yolov8s_custom' # ,imgsz=512 # Replace with the actual file path
)
```

```python
# Validation
from ultralytics import YOLO
model = YOLO('/content/drive/MyDrive/omdena/runs/detect/yolov8s_custom3/weights/best.pt') # Replace with the actual file path
model.val(
   data='/content/drive/MyDrive/omdena/Yolo/datatest.yaml', # Replace with the actual file path
   conf=0.25, save=True
)
```

```python
# Detection on test images (predict)
from ultralytics import YOLO

model = YOLO('/content/drive/MyDrive/omdena/runs/detect/yolov8s_custom3/weights/best.pt') # Replace with the actual file path
model.predict(
   source='/content/drive/MyDrive/omdena/Yolo/test/images', # Replace with the actual file path
   conf=0.25, save=True
)
```
